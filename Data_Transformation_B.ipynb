# Data Cleaning and Manipulation - B

```python
# import required packages
import pandas as pd
```

## Merging all excel sheets into one

```python
# importing the first sheet of excel to just fetch the columns
d1 = pd.read_excel('/content/NBA_Shooting_Totals.xlsx', sheet_name=0)
cols = d1.columns.tolist().append('Season')

# create a new empty df with the above columns
df = pd.DataFrame(columns=cols)
```

```python
# get all the names in the sheet and store it in SEASON variable
l = pd.read_excel("/content/NBA_Shooting_Totals.xlsx", None)
seasons = list(l.keys())
```

```python
# append all the sheets in excel along with the SEASON
for i in range(21):
    dd = pd.read_excel('/content/NBA_Shooting_Totals.xlsx', sheet_name=i)
    dd['SEASON'] = seasons[i]
    df = df.append(dd)

#print(df.describe(include='all'))
#df.to_excel('~/Downloads/data.xlsx', index=False)
```

```python
# df head
df.head()
```

```python
# converting the years in SEASON to full form
df['SEASON'] = df['SEASON'].apply(lambda x: x[:-2]+'19'+x[-2:] if int(x[-2:]) > 80  else x[:-2]+'20'+x[-2:] )
```

```python
# currenlty how season looks
df['SEASON'].unique()
```

```python
df.head()
```

## Column Cleaning

```python
# renaming 'Corener' to 'Corner' in columns
df.columns = df.columns.str.replace('Corener','Corner')

# removing 'Unnamed:' from all columns
df.columns = df.columns.str.replace('Unnamed:\s?\d+', '')
```

```python
# merging columns and first data row
df.columns = df.columns + ' ' + df.iloc[0]
```

```python
# get columns as list
cols = df.columns.to_list()
cols
```

```python
# making manual changes to the column names
cols[4] = 'Restricted Area ' + cols[4]
cols[5] = 'Restricted Area ' + cols[5]

cols[7] = 'In the Paint (Non RA) ' + cols[7]
cols[8] = 'In the Paint (Non RA) ' + cols[8]

cols[10] = 'Mid Range ' + cols[10]
cols[11] = 'Mid Range ' + cols[11]

cols[13] = 'Left Corner 3 ' + cols[13]
cols[14] = 'Left Corner 3 ' + cols[14]

cols[16] = 'Right Corner 3 ' + cols[16]
cols[17] = 'Right Corner 3 ' + cols[17]

cols[19] = 'Corner 3 ' + cols[19]
cols[20] = 'Corner 3 ' + cols[20]

cols[22] = 'Above the Break 3 ' + cols[22]
cols[23] = 'Above the Break 3 ' + cols[23]
```

```python
# getting rid of whitespaces in columns
for i, col  in enumerate(cols):
  if col[0] == ' ':
    cols[i] = col[1:]

cols
```

```python
# assigning the new cleaned columns to the dataframe
df.columns = cols

# removing the first row from the dataframe as it is merged in the columns
df = df[1:]
```

```python
# extra column cleaning
df.columns = df.columns.str.replace('2018-2019','')
df.columns = df.columns.str.strip()
```

```python
# just a quick look at the dataframe
df.head()
```

## Data Joining

```python
free_throw = pd.read_csv('/content/General_Totals_All_Seasons_Cleaned.csv')
free_throw.head()
```

```python
free_throw = free_throw[['PLAYER', 'Free Throw FGM', 'Free Throw FGA', 'Free Throw FG%', 'SEASON', 'TEAM']]
```

```python
free_throw.head()
```

```python
df_merged = pd.merge(df, free_throw,  how='left', on=['PLAYER','SEASON', 'TEAM'])
df_merged.head()
```

## Data Structuring

```python
# melting the dataframe
df_melt = pd.melt(df_merged, id_vars=['PLAYER','TEAM',	'AGE', 'SEASON'], value_name='VALUE')
```

```python
# splitting the column 'variable'
df_melt['STAT'] = df_melt['variable'].apply(lambda x: x[-3:])
df_melt['ZONE'] = df_melt['variable'].apply(lambda x: x[:-3])
```

```python
# just a quick look at dataframe
df_melt.sort_values(by=['PLAYER', 'ZONE']).head(20)
```

```python
# dropping column 'variable' 
df_melt.drop(columns=['variable'], inplace=True)
```

```python
# removing extra whitespaces from the column 'ZONE'
df_melt['ZONE'] = df_melt['ZONE'].str.strip()
```

```python
# un-pivoting the dataframe
df_pivot = pd.pivot(df_melt, index=['PLAYER','TEAM', 'AGE','SEASON','ZONE'], columns='STAT', values='VALUE')
df_pivot.reset_index(inplace=True)
```

```python
# quick look at the current dataframe
df_pivot.head(20)
```

## Exporting the Final dataframe

```python
# exportingthe dataframe to new CSV
df_pivot.to_csv('Cleaned_Shooting_Data_w_FreeThrow.csv', index = False, header=True) 
```
